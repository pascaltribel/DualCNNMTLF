```python
import pandas as pd 

import lightgbm as lgb
from sklearn.multioutput import MultiOutputRegressor

from utils import create_time_series_splits
```

```python
df = pd.read_csv('./preprocessed_data.csv', sep=',', parse_dates=['Datetime'], index_col=0)
# Transform index in Pandas Datetime format
df.index = pd.to_datetime(df['Datetime'], utc=True)
# Drop Datetime column
df.drop(columns=['Datetime'], inplace=True)
# Reverse ordering to go from top to bottom
# Drop rows with missing values
df.dropna(inplace=True)
# Set columns to be excluded
to_remove = list(set(df.columns).difference({'Total Load', 'Most recent forecast','Day-ahead 6PM P10'}))
```

# LGBM univariate - 24 steps ahead
Hypothesis: most recent forecast snapshot is taken at 6pm and used from midnight to 6pm or to midnight. Let's evaluate this hypothesis

```python
#plot df['Total Load'] and df['Day-ahead 6PM forecast']
data = df.iloc[-1000:][['Total Load', 'Day-ahead 6PM P10','Minute', 'Hour']]
# data.index = data['Minute']
import matplotlib.pyplot as plt
plt.figure(figsize=(18, 6))
plt.plot(abs(data['Total Load'] - data['Day-ahead 6PM P10']), label='Absolute Error')

# if x datetime has minutes 0 draw a vertical line 
flag = 0
for i in data.index:
    if (i.minute == 0) and (i.hour == 0):
        flag += 1 
        if flag <= 2:
            plt.axvline(x=i, color='red', linestyle='--', label='0AM')
        else:
              plt.axvline(x=i, color='red', linestyle='--')

    if (i.minute == 0) and (i.hour == 18):
        flag += 1 
        if flag <= 2:
            plt.axvline(x=i, color='blue', linestyle='--', label='6PM')
        else:
            plt.axvline(x=i, color='blue', linestyle='--')

#xticks 90 
plt.xticks(rotation=-90)
plt.legend()
plt.grid()
plt.show()
```

(hard to tell) but sudden changes seems present at midnight but not at 6pm. Therefore we hypothesize that forecasting is applied from midnight to midnight. We will therefore do 30 hour forecast at 6pm and drop the first 6h from the evaluation

```python
excluded_columns = ['Total Load', 'Most recent forecast', 'Most recent P10',
       'Most recent P90', 'Day-ahead 6PM forecast', 'Day-ahead 6PM P10',
       'Day-ahead 6PM P90', 'Week-ahead forecast', 'Year', 'Month', 'Day',
       'Hour', 'Minute', 'Season', 'Lockdown']
```

```python
splits = create_time_series_splits(
    data=df,
    train_size_days=7*4, # 1 month for training (all times)
    test_size_days=7*4, # 1 month for testing (only at 6pm)
    num_splits=10,
    window_size_steps=4*6, # 6 hours window
    exclude_columns=excluded_columns,
    target_column='Total Load Interpolated',
    prediction_horizon_steps=4*30, # 30 hours step ahead forecast
    shifting_steps=1,
    elia_column_to_return='Day-ahead 6PM forecast',
    alignment_times=['18:00']
)
```

```python
split_idx = 3
X_train = splits[split_idx]['X_train']
Y_train = splits[split_idx]['Y_train']
X_test = splits[split_idx]['X_test']
Y_test = splits[split_idx]['Y_test']
ELIA_train = splits[split_idx]['ELIA_train']
ELIA_test = splits[split_idx]['ELIA_test']
```

```python
X_train.shape, Y_test.shape
```

```python
lgb_model = lgb.LGBMRegressor(learning_rate=0.05, n_estimators=50)
model = MultiOutputRegressor(lgb_model)
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
```

```python
Y_pred
```

```python
# column wise MAPE 
mape_lgbm = (abs(Y_pred - Y_test.values) / Y_test.values)
```

```python
ELIA_test.columns = Y_test.columns
```

```python
# column wise MAPE 
mape_elia = (abs(ELIA_test.values - Y_test.values) / Y_test.values)
```

```python
# plot mape 
import matplotlib.pyplot as plt
plt.plot(mape_lgbm.mean(axis=1), label='LGBM MAPE')
plt.plot(mape_elia.mean(axis=1), label='ELIA MAPE')
plt.xlabel('Time')
#xticks each 4 steps 
plt.xticks(range(0, len(mape_lgbm.mean(axis=1)), 4))
plt.xticks(rotation=-90)
plt.ylabel('MAPE')
plt.legend()
```
